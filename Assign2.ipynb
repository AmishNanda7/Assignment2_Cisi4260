```python
import requests
from bs4 import BeautifulSoup
import re

def scrape_article(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "lxml")

        title = soup.title.get_text(strip=True) if soup.title else "No Title"

        paragraphs = soup.find_all("p")
        article_text = " ".join([p.get_text(" ", strip=True) for p in paragraphs])
        article_text = re.sub(r"\s+", " ", article_text).strip()

        return {
            "url": url,
            "title": title,
            "text": article_text,
            "error": ""
        }

    except Exception as e:
        return {
            "url": url,
            "title": "Error",
            "text": "",
            "error": str(e)
        }
```

```python
articles = []

for url in urls:
    result = scrape_article(url)
    articles.append(result)

len(articles)
```

```python
df = pd.DataFrame(articles)
df.head()
```

```python
df = df[df["text"].str.len() > 200]
df = df.reset_index(drop=True)
df.shape
```

```python
stop_words = set(stopwords.words("english"))

def summarize_text(text, max_sentences=2):
    if not text or len(text.strip()) == 0:
        return ""

    sentences = sent_tokenize(text)
    if len(sentences) <= max_sentences:
        return text

    words = word_tokenize(text.lower())
    words = [word for word in words if word.isalnum() and word not in stop_words]

    word_freq = Counter(words)

    sentence_scores = {}
    for sentence in sentences:
        sentence_words = word_tokenize(sentence.lower())
        score = sum(word_freq.get(word, 0) for word in sentence_words)
        sentence_scores[sentence] = score

    top_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:max_sentences]

    summary = " ".join(top_sentences)
    return summary
```

```python
df["summary"] = df["text"].apply(summarize_text)
df[["title", "summary"]].head()
```

```python
topic_keywords = [
    "ai", "artificial intelligence", "education", "students",
    "learning", "teaching", "classroom", "university", "school"
]
```

```python
def calculate_importance_score(text):
    if not text or len(text.strip()) == 0:
        return 0

    # Sentiment score
    polarity = TextBlob(text).sentiment.polarity

    # Topic relevance score
    text_lower = text.lower()
    keyword_count = sum(text_lower.count(keyword) for keyword in topic_keywords)

    # Combine them
    score = polarity + (keyword_count * 0.05)

    return round(score, 3)
```

```python
def get_direction(score):
    if score >= 0:
        return "Positive"
    else:
        return "Negative"
```

```python
df["importance_score"] = df["text"].apply(calculate_importance_score)
df["direction"] = df["importance_score"].apply(get_direction)
```

```python
final_df = df[["url", "title", "summary", "importance_score", "direction"]]
final_df.head()
```
